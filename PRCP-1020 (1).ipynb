***Missing Values***


# Here we will check the pecentage of non values present in each feature
# 1- features which has missing value
# 2- features name and percentage of missing values.


```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in features_with_na:
  print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing values')
```

Let's plot some diagram for this relationship

```python
for feature in features_with_na:
  data=dataset.copy()
  data[feature]=np.where(data[feature].isnull(),1,0)
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

Here With the relation the missing values and the dependent variable is clearly visible. So we need to replace these nan values with something meaningful which we will do in the Feature Engineering Section.

**Numerical Variables**

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype!='0']
print("Number of numerical variables:",len(numerical_features))
dataset[numerical_features].head()
```

```python
# list of variable that contains year information
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
#Analyze the Temporal Datetime Variables
# we will check wheather there is a relation between yer the house is sold and SalePrice
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs YearSold
```

```python
year_feature
```

```python
# Here we will compare the difference between All years feature with SalePrice
for feature in year_feature:
  if feature!='YrSold':
    data=dataset.copy()
    data[feature]=data['YrSold']-data[feature]

    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
# Numerical Variables are usually of 2 type
# Continous and discrete variables
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
# Lets Find the relationship between them and SalePrice

for feature in discrete_feature:
  data=dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('SalePrice')
  plt.title(feature)
  plt.show()
```

**Continuous Variable**

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count: {}".format(len(continuous_feature)))
```

```python
# Lets analyse the continuous values by creating histograms to understands the data

for feature in continuous_feature:
  data=dataset.copy()
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel('Count')
  plt.title(feature)
  plt.show()
```
