```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("A_data.csv")
df.head()
```

```python
## Analyzing Change in Price Over Time

def plot_stock_price(df, ticker):
    plt.figure(figsize=(10, 5))
    plt.plot(df['date'], df['close'], label=f'{ticker} Close Price')
    plt.xlabel('date')
    plt.ylabel('close Price')
    plt.title(f'{ticker} Stock Price Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_stock_price(df, 'A')
```

```python
## Analyzing Moving Averages
```

```python
def plot_moving_averages(df, ticker, window_sizes=[20, 50]):
    plt.figure(figsize=(10, 5))
    plt.plot(df['date'], df['close'], label='Close Price')
    
    for window in window_sizes:
        df[f'MA_{window}'] = df['close'].rolling(window=window).mean()
        plt.plot(df['date'], df[f'MA_{window}'], label=f'{window}-Day MA')
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'{ticker} Moving Averages')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_moving_averages(df, 'A')
```

```python
df1= pd.read_csv("AAPL_data.csv")
df1.head()
```

```python
# 2. Analyze Price Change
df1['Daily Change %'] = df1['close'].pct_change() * 100

# 3. Moving Averages
df1['SMA_20'] = df1['close'].rolling(window=20).mean()
df1['SMA_50'] = df1['close'].rolling(window=50).mean()

# Plotting
plt.figure(figsize=(12,6))
plt.plot(df1['close'], label='Close Price')
plt.plot(df1['SMA_20'], label='20-Day SMA')
plt.plot(df1['SMA_50'], label='50-Day SMA')
plt.title('AAPL Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
## Analyzing Change in Price Over Time

def plot_stock_price(df1, ticker):
    plt.figure(figsize=(10, 5))
    plt.plot(df1['date'], df1['close'], label=f'{ticker} Close Price')
    plt.xlabel('date')
    plt.ylabel('close Price')
    plt.title(f'{ticker} Stock Price Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_stock_price(df1, 'AAPL')
```

```python
# 1. Calculate daily change in closing price and percent
df1['close Change'] = df1['close'].diff()
df1['close Change %'] = df1['close'].pct_change() * 100

# 2. Plot Closing Price
plt.figure(figsize=(12,6))
plt.plot(df1['close'], label='AAPL Closing Price', color='blue')
plt.title('Apple (AAPL) Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
