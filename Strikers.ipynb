# Statistical Analysis:

```python
# Determine which nationality strikers have the highest average number of goals scored. 
avg_goals_by_nationality = data.groupby("Nationality")["Goals Scored"].mean()
highest_avg_goals_nationality = avg_goals_by_nationality.idxmax()
highest_avg_goals_value = avg_goals_by_nationality.max()

print(f"The nationality with the highest average goals scored is {highest_avg_goals_nationality} with {highest_avg_goals_value:.2f} goals on average.")
```

```python
# Calculate the average conversion rate for players based on their footedness. 
# assignment Ques 2 is What is the portion of Right-footed strikers within the dataset? 
# assignment Ques 4 is What is the average conversion rate for left-footed player? 
# Average conversion rate for each footedness type

avg_conversion_rate = data.groupby("Footedness")["Conversion Rate"].mean()

print("Average Conversion Rate by Footedness:",avg_conversion_rate)
```

```python
# Find whether there is any significant difference in consistency rates among strikers from various nationalities.
# Before doing the appropriate test, must check for the assumptions. 

from scipy import stats
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")


# Step 1: Nationality with the highest average goals scored
avg_goals_by_nationality = data.groupby('Nationality')['Goals Scored'].mean().sort_values(ascending=False)
highest_avg_goals = avg_goals_by_nationality.head(1)
print("\nHighest Average Goals by Nationality:\n", highest_avg_goals)

# Step 2: Average conversion rate by footedness
avg_conversion_rate_by_footedness = data.groupby('Footedness')['Conversion Rate'].mean()
print("\nAverage Conversion Rate by Footedness:\n", avg_conversion_rate_by_footedness)
#Find whether there is any significant difference in consistency rates among strikers from various nationalities. Before doing the appropriate test, must check for the assumptions. 
# Step 3: ANOVA to test difference in consistency rates among nationalities
# First, check assumptions for ANOVA:
# 1. Normality (Shapiro-Wilk Test)
normality_test = stats.shapiro(data['Consistency'])
print("\nShapiro-Wilk Test for Normality of Consistency:\n", normality_test)

# 2. Homogeneity of variances (Levene's Test)
levene_test = stats.levene(*[group['Consistency'].values for name, group in data.groupby('Nationality')])
print("\nLevene's Test for Homogeneity of Variance:\n", levene_test)


# If assumptions are met, perform ANOVA:
anova_test = stats.f_oneway(*[group['Consistency'].values for name, group in data.groupby('Nationality')])
print("\nANOVA Test Result:\n", anova_test)

# Step 4: Correlation between hold-up play and consistency rate
# First, check for normality
shapiro_holdup = stats.shapiro(data['Hold-up Play'])
shapiro_consistency = stats.shapiro(data['Consistency'])

if shapiro_holdup.pvalue > 0.05 and shapiro_consistency.pvalue > 0.05:
    correlation = stats.pearsonr(data['Hold-up Play'], data['Consistency'])
    print("\nPearson's Correlation (Normal Distribution):\n", correlation)
else:
    correlation = stats.spearmanr(data['Hold-up Play'], data['Consistency'])
    print("\nSpearman's Correlation (Non-Normal Distribution):\n", correlation)

# Step 5: Regression analysis to see if hold-up play influences consistency
X = data['Hold-up Play']
X = sm.add_constant(X)  # Add intercept
y = data['Consistency']

# Fit the regression model
model = sm.OLS(y, X).fit()

print("\nRegression Analysis:\n", model.summary())
plt.figure(figsize=(8, 6))
sns.barplot(x='Hold-up Play', y='Consistency', data=data, palette='coolwarm')
plt.title('Effect of Hold-up Play on Consistency')
plt.show()
```

# Feature Engineering: 

```python
# Create a new feature - Total contribution score by summing up specific columns: 'Goals Scored', 'Assists', 
# 'Shots on Target', 'Dribbling Success', 'Aerial Duels Won', 'Defensive Contribution', 'Big Game Performance', 'Consistency'.

# Define the columns to sum up
columns_to_sum = [
    "Goals Scored", "Assists", "Shots on Target", "Dribbling Success", 
    "Aerial Duels Won", "Defensive Contribution", "Big Game Performance", "Consistency"
]

# Ensure all columns exist in the dataframe
columns_to_sum = [col for col in columns_to_sum if col in data.columns]

# Convert to numeric (handle errors) and fill missing values with 0
data[columns_to_sum] = data[columns_to_sum].apply(pd.to_numeric, errors='coerce').fillna(0)

# Create the new feature "Total Contribution Score"
data["Total Contribution Score"] = data[columns_to_sum].sum(axis=1)

# Display the first few rows with the new feature
print(data[["Total Contribution Score"]].head())

```

```python
# Encode the Footedness and marital status by LabelEncoder.

from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
data['Footedness']=encoder.fit_transform(data['Footedness'])
data['Marital Status']=encoder.fit_transform(data['Marital Status'])
data.head()
```

```python
# Create dummy variables for Nationality and add them to the data.

dummies=pd.get_dummies(data['Nationality'])
processed_data=pd.concat([data,dummies],axis=1)
processed_data=processed_data.drop('Nationality',axis=1)
processed_data.head()
```

```python
Nationality_encoder=LabelEncoder()
processed_data['Brazil']=Nationality_encoder.fit_transform(processed_data['Brazil'])
processed_data['England']=Nationality_encoder.fit_transform(processed_data['England'])
processed_data['France']=Nationality_encoder.fit_transform(processed_data['France'])
processed_data['Germany']=Nationality_encoder.fit_transform(processed_data['Germany'])
processed_data['Spain']=Nationality_encoder.fit_transform(processed_data['Spain'])
     
```

# Clustering Analysis: 

```python
# Perform KMeans clustering: 

### Select features by dropping the Striker_ID from the updated data.
### Calculate the Within-Cluster-Sum-of-Squares (WCSS). 
### Visualize the elbow chart to select the optimal number of clusters (The breakpoint of elbow chart must show 2). 

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Initialize WCSS list
wcss = []

# Prepare the data
x = processed_data.drop('Striker_ID', axis=1)

# Compute WCSS for cluster counts from 1 to 14
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=1, random_state=42)
    kmeans.fit(x)
    wcss_score = kmeans.inertia_
    wcss.append(wcss_score)

# Plot the Elbow Method
plt.plot(range(1, 15), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Build the KMeans cluster with the optimal number of clusters and add the labels into the data.  
final_kmeans = KMeans(n_clusters = 2,n_init=10)
final_kmeans.fit(x)

# Generating labels
labels = final_kmeans.labels_
labels
     
```

```python
processed_data['cluster']=labels
processed_data.head()
     
```

```python
# Calculate the average total contribution score by the value of clusters. 

processed_data.groupby('cluster')[['Total Contribution Score']].mean()
```

```python
# Assign the tag 'Best strikers' for 0 and 'Regular strikers' for 1 and add a new column 'Strikers types' into the data. Drop the Clusters variable. 

# Map cluster values to Striker Types
striker_types = {0: "Best strikers", 1: "Regular strikers"}
processed_data["Strikers Types"] = processed_data["cluster"].map(striker_types)

# **Drop the Cluster column**
processed_data = processed_data.drop(columns=["cluster"], errors="ignore")
```

```python
processed_data.head()
```

```python
# Use feature mapping to map the new feature Strikers types: 'Best strikers' for 1 and 'Regular strikers' for 0. 

striker_type_mapping = {"Best strikers": 1, "Regular strikers": 0}
processed_data["Strikers Types"] = processed_data["Strikers Types"].map(striker_type_mapping)

```

# Machine Learning Model: 

```python
processed_data=processed_data.drop("Striker_ID", axis= 1)
processed_data.head()
```

```python
# Select the features into x and the target column Strikers types into y.
# Must delete unnecessary columns (i.e., 'Strikers_ID') while selecting the features. 

# Selecting Features

x = processed_data.drop(['Strikers Types'], axis = 1)
y = processed_data['Strikers Types']
```

Scaling features

```python
# Perform feature scaling with StandardScaler and split the data into train and test sets where the test data size will be 20%

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaled=scaler.fit_transform(x)
scaled
```

Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(scaled,y,test_size=0.2,random_state=42)
```

# Predictive Classification Analystics

```python
# Build a logistic regression machine learning model to predict strikers type.
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix

# Make predictions and evaluate by calculating the accuracy percentage.

model= LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy*100,'%')
```

```python
# Create the confusion matrix and visualize it

conf_matrix=confusion_matrix(y_test,y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix,annot=True,fmt="d",cmap="Blues",annot_kws={"size": 14},cbar=False)
plt.title('Confusion matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
